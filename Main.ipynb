```
#loan amount distribution for repaid projects 
loanamount_repaid <- data %>% filter(!is.na(repaid)) %>% group_by(repaid) %>% summarise(avg_loan = mean(loan_amount_asked))
loanamount_repaid
ggplot(loanamount_repaid, aes(x = factor(repaid), y = avg_loan)) +  geom_bar(stat = "identity", fill = "darkblue") +
  geom_text(aes(label = round(avg_loan, 2)), vjust = -0.5, color = "black", size = 4) +
  theme_minimal()
```

```
#Boxplot for the distribution of the loan amount asked
ggplot(data, aes(x = factor(repaid), y = loan_amount_asked)) +
  geom_boxplot(fill = "skyblue")
```

```
data1 <- data %>% filter(!is.na(repaid))
cor(data1$boostingbid_bythefirm, data1$repaid)
```

```
#pecentages of repaid for the presentation
cont_table <- table(data1$funded, data1$repaid)
percentages <- prop.table(cont_table, margin = 1) * 100
percentages
```

```
data %>% filter(!is.na(repaid)) %>% group_by(boostingbid_bythefirm) %>% summarise(n_proj=n(), repaid_rate=sum(repaid == 1)/n())
```

```
data %>% filter(!is.na(repaid)) %>% group_by(provider) %>% summarise(n_proj=n(), repaid_rate=sum(repaid == 1)/n())
```

```
#Graph for Projects per provider
provider_projects <- data %>% filter(!is.na(repaid)) %>% group_by(provider) %>% summarise(n_proj=n(), )

ggplot(provider_projects, aes(x = factor(provider), y = n_proj)) + 
    geom_histogram(stat = "identity", fill = "skyblue")
```

```
providers_role <- data %>% filter(!is.na(repaid)) %>% group_by(provider) %>% summarise(repaid_rate=sum(repaid == 1)/n())

#Graph for repayment rate per provider
ggplot(providers_role, aes(x = repaid_rate, y = factor(provider))) +
  geom_histogram(stat = "identity", fill = "blue")
```

```
data$interest_difference <- data$borrower_max_int - data$final_avg_int_rate
```

```
sector_df= data%>%
    group_by(sect_fishingfarming, sect_manufacturing, sect_services, sect_rawmaterials)%>%
    summarise(avg_interest_rate = mean(final_avg_int_rate, na.rm = TRUE), count= n(),n_repaid = sum(repaid == 1, na.rm = TRUE))%>%
    rowwise() %>%
    mutate(proportion= n_repaid/count, num_sectors = sum(sect_fishingfarming, sect_manufacturing, sect_services, sect_rawmaterials))

sector_df 
```

```
sum_sector <- sector_df %>%
  group_by(num_sectors) %>%
  summarise(
    avg_interest_rate = mean(avg_interest_rate),
    min_interest_rate = min(avg_interest_rate),
    max_interest_rate = max(avg_interest_rate),
    avg_repaid_proportion = mean(proportion),
    count= sum(count),
    num_combinations = n()

  )
sum_sector
```
